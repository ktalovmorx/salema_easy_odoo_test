```python
/** @odoo-module **/

import { Navbar } from "@point_of_sale/app/navbar/navbar";
import { patch } from "@web/core/utils/patch";
import { useService } from "@web/core/utils/hooks";

/* ¿ Que es esto de Patch, para que sirve ? */
patch(Navbar.prototype, {
    setup() {
        super.setup(...arguments);
        this.orm = useService("orm");
        this.notificationService = useService("notification");
        this.rpc = useService("rpc");
        this.retail_type = "";
        this._loadRetailType();
    },

    /* ¿ A que se refiere esto de async ? */
    async _loadRetailType() {
        try {
            const result = await this.rpc("/jm_pdv/get_retail_type", {});
            this.retail_type = result.retail_type || "";
            this.render();
        } catch (error) {
            console.error("Error cargando retail_type:", error);
        }
    },
    getRetailType() {
        return this.retail_type;
    },
    go_home() {
        window.open("/", "_blank");
    },
    open_comandas() {
        window.open("/comandas", "_blank");
    },
    async delete_floating_orders() {
        try {
            const confirmDelete = window.confirm("¿Está seguro de que desea eliminar las órdenes flotantes?");
            if (!confirmDelete) return;

            /*¿ Que esta haciendo esta linea? */
            const result = await this.orm.call("pos.order", "delete_floating_orders", []);

            /*¿ Cual seria la funcionalidad de esta linea? */
            this.notificationService.add("Notificación", {
                title: result.message || "Órdenes flotantes eliminadas",
                type: result.status || "success",
            });
        } catch (error) {
            this.notificationService.add("Error", {
                title: error.message || "No se pudieron eliminar las órdenes flotantes",
                type: "danger",
            });
        }
    },
    open_products_backend() {
        window.open("/web#action=point_of_sale.product_template_action_pos_product", "_blank");
    },
    refresh_page() {
        window.location.reload();
    },
});
```
